```python
# Step 1: Import libraries
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: Set seaborn theme
plt.style.use('seaborn-v0_8')  # Updated style for matplotlib

# Step 2: Download Apple stock data for the last 2 years
apple_data = yf.download("AAPL", period="2y", interval="1d")

# Step 3: Basic data inspection
apple_data.reset_index(inplace=True)  # Move Date from index to column
print(apple_data.head())

# Step 4: Line Chart of Closing Price
plt.figure(figsize=(14, 6))
plt.plot(apple_data['Date'], apple_data['Close'], label='Close Price', color='blue')
plt.title('Apple Stock Closing Price (Last 2 Years)')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Step 5: Moving Average (50-day)
apple_data['MA50'] = apple_data['Close'].rolling(window=50).mean()
```

```python
correlation_data = apple_data[['Open', 'High', 'Low', 'Close', 'Volume']]
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_data.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Between Apple Stock Price Features')
plt.tight_layout()
plt.show()
```
